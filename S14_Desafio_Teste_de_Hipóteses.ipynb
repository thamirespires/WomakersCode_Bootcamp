```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import ttest_1samp
```

```python
df = pd.read_csv('experimento_teste_ab.csv')
df.head()
```

**1. Qual dos cenários tem a maior taxa de conversão?**

```python
# calcular a taxa de conversão para a versão A
total_conversoes_A = df[df['Versão_Página'] == 'A']['Conversões'].sum()
total_visualizacoes_A = df[df['Versão_Página'] == 'A']['Visualizações'].sum()
taxa_conversao_A = (total_conversoes_A / total_visualizacoes_A) * 100

# calcular a taxa de conversão para a versão B
total_conversoes_B = df[df['Versão_Página'] == 'B']['Conversões'].sum()
total_visualizacoes_B = df[df['Versão_Página'] == 'B']['Visualizações'].sum()
taxa_conversao_B = (total_conversoes_B / total_visualizacoes_B) * 100

print('taxa de conversão A:', taxa_conversao_A, "%")
print('taxa de conversão B:', taxa_conversao_B, "%")
```

*Podemos analisar que o cenário B tem a maior taxa de conversão*

**2. Calcule qual o tamanho da amostra necessária para o desenvolvimento
de um teste A/B, seguindo os seguintes critérios:**

* a. O cenário A, da base, como o inicial, que funciona hoje.
* b. Considere 95% de confiança de que o efeito na conversão não foi
aleatório com um nível de significância de 5% (alpha = 0,05).
* c. Também considere 80% de certeza conseguir capturar o efeito
da nova abordagem.
* d. O aumento para 10% de conversão.

```python
taxa_conversao_alvo = 0.10 #10%
delta = taxa_conversao_alvo - taxa_conversao_A

# calcular a variavel binomial para cada observação no grupo controle
observacoes_controle = df[df['Versão_Página'] == 'A']['Conversões'].tolist()
# calcular a variancia amostral do grupo controle
variancia_amostral_A = np.var(observacoes_controle)

alpha = 0.05
beta = 1 - 0.80
z_alpha_2 = stats.norm.ppf(1 - alpha/2)
z_beta = stats.norm.ppf(1-beta)
#calculando o tamanho da amostra necessario
n= (2* (variancia_amostral_A)*(z_alpha_2 + z_beta )**2) / delta**2

print("tamanho da amostra necessario em cada grupo:", round(n))
```

**3. Considerando uma amostra de 45 números que representam o index do
dataframe, índices= ([909, 751, 402, 400, 726, 39, 184, 269, 255, 769,
209, 715, 677, 381, 793, 697, 89, 280, 232, 756, 358, 36, 439, 768, 967,

que a idade média das pessoas com problemas cardíacos é maior que
50 anos? Nível de significância igual a 5%.**

```python
pacientes = pd.read_csv('pacientes.csv', index_col='Unnamed: 0')
pacientes.head()
```

```python
indices_amostra = [909, 751, 402, 400, 726, 39, 184, 269, 255, 769, 209, 715, 677, 381, 793, 697, 89, 280, 232, 756, 358, 36, 439, 768, 967, 699, 473, 222, 89, 639, 883, 558, 757, 84, 907, 895, 217, 224, 311, 348, 146, 505, 273, 957, 362]

amostra_idades =  pacientes.loc[indices_amostra]["Idade"]

_, p_value = ttest_1samp (amostra_idades,50)

alpha = 0.05

if p_value < alpha:
  print( 'rejeitamos a hipotese nula')
else:
  print(' não rejeitamos a hipotese nula')
```

**4. Queremos entender que tipo de amostra estamos lidando se dividirmos
os conjuntos em 2, sendo um com pessoas que têm condições de
saúde adicionais e o outro com pessoas saudáveis. Seria dependente
ou independente?**

*Se você dividir o conjunto de dados em dois grupos com base no estado de saúde
(um grupo para pessoas com condições de saúde adicionais e outro grupo para pessoas saudáveis), a natureza da amostra seria independente.
 Isso ocorre porque os dois grupos seriam formados de forma independente um do outro, ou seja, a inclusão de um indivíduo em um grupo não afetaria a inclusão de outro indivíduo no mesmo grupo.*

*Portanto, a caracterização da amostra como independente é apropriada neste contexto.*

**5. Agora considere o um conjunto de pessoas aleatória que representam o index do dataframe, índices = ([690, 894, 67, 201, 364, 19, 60, 319, 588, 643, 855, 623, 530, 174, 105, 693, 6, 462, 973, 607, 811, 346, 354, 966, 943, 372]), podemos dizer que a pressão arterial média para pacientes com condições de saúde adicionais é igual à pressão arterial média para pacientes sem condições adicionais de saúde? Considere o nível de significância a 6%.**

```python
from scipy.stats import ttest_ind

# Definindo os índices dos pacientes com condições adicionais de saúde
indices_com_condicoes = [690, 894, 67, 201, 364, 19, 60, 319, 588, 643, 855, 623, 530, 174, 105, 693, 6, 462, 973, 607, 811, 346, 354, 966, 943, 372]

# Filtrando o DataFrame para incluir apenas os pacientes com condições adicionais de saúde
parcial = pacientes.loc[indices_com_condicoes]

# Separando as pressões arteriais dos pacientes com e sem condições adicionais de saúde
pressao_com_condicoes_adicionais = parcial[parcial['Nome_Estado_Saude'] != 'Saudável']['Pressao_Arterial']
pressao_sem_condicoes_adicionais = parcial[parcial['Nome_Estado_Saude'] == 'Saudável']['Pressao_Arterial']

# Realizando o teste t de independência
statistic, p_value = ttest_ind(pressao_com_condicoes_adicionais, pressao_sem_condicoes_adicionais)
print("Estatística do teste t:", statistic)
print("Valor de p:", p_value)

# Definindo o nível de significância
alpha = 0.06

# Verificando se rejeitamos ou não a hipótese nula
if p_value < alpha:
    print("Rejeitamos a hipótese nula")
else:
    print("Não rejeitamos a hipótese nula")
```

*Como não rejeitamos a hipótese nula, isso significa que, com base na amostra analisada e no nível de significância escolhido (6%), não há evidências estatísticas suficientes para afirmar que a pressão arterial média para pacientes com condições adicionais de saúde é diferente da pressão arterial média para pacientes sem condições adicionais de saúde.*

**6. Existe uma diferença significativa na pressão arterial média entre
diferentes grupos étnicos nesta população? (Teste ANOVA, alpha é 5%)**

* a. Hipótese Nula (H0): A pressão arterial média é a mesma em todos
os grupos étnicos.
* b. Hipótese Alternativa (H1): Há uma diferença significativa na
pressão arterial média entre pelo menos dois grupos étnicos
nesta população.

```python
from scipy.stats import f_oneway

pressao_por_etnia = {
    "1": pacientes[pacientes['Nome_Etnia'] == "Afro-americano"]['Pressao_Arterial'].values,
    "2": pacientes[pacientes['Nome_Etnia'] == "Asiático"]['Pressao_Arterial'].values,
    "3": pacientes[pacientes['Nome_Etnia'] == "Hispânico"]['Pressao_Arterial'].values,
    "4": pacientes[pacientes['Nome_Etnia'] == "Caucasiano"]['Pressao_Arterial'].values,
  }

statistic, p_value = f_oneway(*pressao_por_etnia.values())

alpha = 0.05

if p_value < alpha:
    print("Rejeitamos a hipótese nula")
else:
    print("Não rejeitamos a hipótese nula")
```

*Como não rejeitamos a hipótese nula neste teste ANOVA, isso significa que, com base na amostra analisada e no nível de significância escolhido (5%), não há evidências estatísticas suficientes para afirmar que há uma diferença significativa na pressão arterial média entre os diferentes grupos étnicos nesta população.*

**7. Lúcia é uma pesquisadora e tem o objetivo de entender a relação de
gênero neste grupo de pacientes. Acredita-se que há uma relação
entre o sexo e condições de saúde adicionais. (Teste qui-quadrado)**

```python
frequencia_condicoes_adicionais_para_homem = pacientes[(pacientes['Nome_Estado_Saude'] != 'Saudável') & (pacientes['Nome_Genero'] == 'Masculino')]
frequencia_sem_condicoes_adicionais_para_homem = pacientes[(pacientes['Nome_Estado_Saude'] == 'Saudável') & (pacientes['Nome_Genero'] == 'Masculino')]
frequencia_condicoes_adicionais_para_mulher = pacientes[(pacientes['Nome_Estado_Saude'] != 'Saudável') & (pacientes['Nome_Genero'] == 'Feminino')]
frequencia_sem_condicoes_adicionais_para_mulher = pacientes[(pacientes['Nome_Estado_Saude'] == 'Saudável') & (pacientes['Nome_Genero'] == 'Feminino')]
```

```python
from scipy.stats import chi2_contingency

tabela_contingencia = [
    [frequencia_condicoes_adicionais_para_homem.shape[0], frequencia_sem_condicoes_adicionais_para_homem.shape[0]],
    [frequencia_condicoes_adicionais_para_mulher.shape[0], frequencia_sem_condicoes_adicionais_para_mulher.shape[0]]
]

statistic, p_value, dof, expected = chi2_contingency(tabela_contingencia)

alpha = 0.05

if p_value < alpha:
    print("Rejeitamos a hipótese nula")
else:
    print("Não rejeitamos a hipótese nula")
```

*Como não rejeitamos a hipótese nula neste teste qui-quadrado, isso significa que não há evidências estatísticas suficientes para afirmar que há uma associação significativa entre o gênero dos pacientes e a presença de condições adicionais de saúde.*

**8. Existe uma associação entre a idade dos pacientes e sua pressão
arterial?**

* a. Hipótese Nula (H0): se a pressão arterial é independente da idade
* b. Hipótese Alternativa (H1): a pressão arterial esta associada a
idade

```python
idades = pacientes['Idade']
pressoes_arteriais = pacientes['Pressao_Arterial']

faixas_etarias = [30, 40, 50, 60, 70, 80]
intervalos_pressao = [100, 120, 140, 160, 180, 200]

# Contar o número de pacientes em cada faixa etária e intervalo de pressão arterial
contagem_idades = np.histogram(idades, bins=faixas_etarias)[0]
contagem_pressao = np.histogram(pressoes_arteriais, bins=intervalos_pressao)[0]

# Criar tabela de contingência
tabela_contingencia = np.array([contagem_idades, contagem_pressao])

# Realizar o teste qui-quadrado de independência
statistic, p_value, dof, expected = chi2_contingency(tabela_contingencia)
print(chi2_contingency(tabela_contingencia))
alpha = 0.05

if p_value < alpha:
    print("Rejeitamos a hipótese nula")
else:
    print("Não rejeitamos a hipótese nula")
```

*Como rejeitamos a hipótese nula neste teste qui-quadrado, isso significa que há evidências estatísticas suficientes para afirmar que há uma associação significativa entre a idade dos pacientes e sua pressão arterial.*

**9. Qual é o intervalo de confiança para a média da pressão arterial entre
os pacientes com condições de saúde adicionais? (nível de confiança
95%)**

```python
from scipy.stats import norm

pressoes_arteriais_condicoes = parcial[parcial['Nome_Estado_Saude'] != 'Saudável']['Pressao_Arterial']

media_pressao_condicoes = np.mean(pressoes_arteriais_condicoes)
desvio_padrao_pressao_condicoes = np.std(pressoes_arteriais_condicoes)

n = len(pressoes_arteriais_condicoes)

nivel_confianca = 0.95

z = norm.ppf((1 + nivel_confianca)/2)

#calcular o intervalo de confinça
intervalo_confianca = z * (desvio_padrao_pressao_condicoes / np.sqrt(n))

limite_inferior = media_pressao_condicoes - intervalo_confianca
limite_superior = media_pressao_condicoes + intervalo_confianca

print('limite inferior',limite_inferior )
print('limite superior',limite_superior)
```

**10. A distribuição da pressão arterial na população segue uma distribuição
normal?**

* a. Hipótese Nula (H0): A distribuição da pressão arterial na
população segue uma distribuição normal.
* b. Hipótese Alternativa (H1): A distribuição da pressão arterial na
população não segue uma distribuição normal.

```python
from scipy.stats import kstest, norm

pressao_arterial_populacao = pacientes['Pressao_Arterial']

# Realizar o teste KS
statistic, p_value = kstest(pressao_arterial_populacao, 'norm')
print("Estatística do teste KS:", statistic)
print("Valor de p:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Rejeitamos a hipótese nula")
else:
    print("Não rejeitamos a hipótese nula")
```

*Como rejeitamos a hipótese nula neste teste KS, isso significa que há evidências estatísticas suficientes para afirmar que a distribuição da pressão arterial na população não segue uma distribuição normal.*
