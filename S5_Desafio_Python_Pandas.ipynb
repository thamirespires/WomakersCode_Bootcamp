```python
# importando bibliotecas
import pandas as pd
import numpy as np
```

```python
# lendo o data frame
df = pd.read_csv("saude_do_sono_estilo_vida.csv")
```

**1. Ao visualizar a base você percebeu que seria melhor alterar o nome de algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’, a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.**

```python
df.rename(columns={'ID': 'Identificador',
                   'Ocupação': 'Profissão',
                   'Categoria BMI': 'Categoria IMC',
                   'Pressão sanguíneaaaa': 'Pressão sanguínea'}, inplace=True)
```

**2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]**

```python
estatisticas_sono = df.groupby('Profissão')['Duração do sono'].agg([np.mean, np.median, pd.Series.mode])
print(estatisticas_sono)
```

**3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?**

```python
eng_soft_df = df[df['Profissão'] == 'Eng. de Software']
total_eng_soft = len(eng_soft_df)
num_obesos = len(eng_soft_df[eng_soft_df['Categoria IMC'] == 'Obesidade'])
porcentagem_obesos = (num_obesos / total_eng_soft) * 100
print(f'A porcentagem de obesos entre as pessoas que atuam com Engenharia de Software é de: {porcentagem_obesos:.2f}%')
```

**4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)**

```python
# Calcula a média de duração do sono para todas as profissões
media_por_profissao = df.groupby('Profissão')['Duração do sono'].mean()

print("Média de duração do sono por profissão:")
print(media_por_profissao)
```

*Ao analisar individualmente a média de sono de cada profissão é possivel dizer que ser advogado faz você dormir bastante em torno de 7.4 horas por noite, sendo o maior valor de todas as profissões, mas ser representande de vendas, faz o individuo dormir bem menos em torno de 5.9 horas por noite, sendo a profissão que mais sofre com menos horas de sono*

**5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)**

```python
enfermeiro_medico_df = df[df['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])]
media_enfermeiro_medico = enfermeiro_medico_df.groupby('Profissão')['Duração do sono'].mean()

print("Média de duração do sono para enfermeiros e médicos:")
print(media_enfermeiro_medico)
```

*Com base no resultado, podemos dizer que o médico(a) tem menos horas de sono em relação ao enfermeiro com 6.9 horas por noite*

**8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)**

```python
media_pressao_homens = df[df['Gênero'] == 'Homem']['Pressão sanguínea'].apply(lambda x: int(x.split('/')[0])).mean()
media_pressao_mulheres = df[df['Gênero'] == 'Mulher']['Pressão sanguínea'].apply(lambda x: int(x.split('/')[0])).mean()

print(f"Média da pressão sanguínea para homens: {media_pressao_homens}")
print(f"Média da pressão sanguínea para mulheres: {media_pressao_mulheres}")
```

*Analisando os resultados podemos dizer que as mulheres tem maior pressão sanguinea com um valor de 130,14*

**10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)**

```python
media_passos_acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_70_ou_menos = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

print("Média menos de 70 :", media_passos_70_ou_menos)
print("Média mais de 70 :", media_passos_acima_70)
```

*Podemos analisar que pessoas com frequencia cardiaca menor de 70 dão mais passos que pessoas com frequencia cardiaca maior de 70*
